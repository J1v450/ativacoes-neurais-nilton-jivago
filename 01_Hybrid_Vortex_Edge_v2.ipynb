## 🌪️ Hybrid Vortex Edge v2

Curva com aceleração nas bordas e suavidade no centro.

```python
import numpy as np
import matplotlib.pyplot as plt

def vortex_edge_v2(x):
    return np.tanh(x) * (1 - np.exp(-np.abs(x)))

x = np.linspace(-5, 5, 400)
y = vortex_edge_v2(x)

plt.plot(x, y)
plt.title("Hybrid Vortex Edge v2")
plt.grid(True)
plt.show()
```
