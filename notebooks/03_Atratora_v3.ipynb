## 🧲 Atratora v3

Curva com modulação central e derivadas ativas.

```python
import numpy as np
import matplotlib.pyplot as plt

def atratora_v3(x):
    return np.tanh(x) * (1 - np.exp(-x**2))

x = np.linspace(-5, 5, 400)
y = atratora_v3(x)

plt.plot(x, y)
plt.title("Atratora v3")
plt.grid(True)
plt.show()
```
