## 📈 Tanh Escalada

Tangente hiperbólica escalada para [0, 1].

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh_escalada(x):
    return (np.tanh(x) + 1) / 2

x = np.linspace(-5, 5, 400)
y = tanh_escalada(x)

plt.plot(x, y)
plt.title("Tanh Escalada")
plt.grid(True)
plt.show()
```
